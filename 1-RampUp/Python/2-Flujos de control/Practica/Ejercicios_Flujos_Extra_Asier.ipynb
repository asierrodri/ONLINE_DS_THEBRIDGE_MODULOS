## Ejercicio Extra
Programa una calculadora que

* Pregunte qué operación quieres realizar (multiplicación/división/suma/resta/potencia/salir)
* Pida dos números 
* Realice la operación
* Siga funcionando hasta que el usuario elija salir

#### Bonus:
* Que guarde todas las operaciones realizadas y las imprima por pantalla cuando el usuario elija esa opción
* Que sea robusta frente a entradas erróneas del usuario o frente a operaciones como dividir por 0

```python
import math
import pprint
import time
```

```python
salir = ""
historial = []
entrada_valida = False

while salir != "salir":
    usuario_operacion = input("¿Qué operación desea realizar?").lower().strip().replace("ó", "o")
    #usuario_operacion = usuario_operacion.strip()
    #usuario_operacion = usuario_operacion.replace("ó", "o")

    #Multiplicación
    if usuario_operacion == "multiplicacion" or usuario_operacion == "multiplicar" or usuario_operacion == "x" or usuario_operacion == "*":
        while not entrada_valida:
            try:
                numero1 = input("Introduce el primer número a multiplicar")
                numero1 = float(numero1.replace(",", "."))
                numero2 = input("Introduce el segundo número a multiplicar")
                numero2 = float(numero2.replace(",", "."))
                entrada_valida = True
            except Exception:
                pprint("Por favor introduzca un número válido")

        entrada_valida = False
        resultado = numero1*numero2
        print(f"Resultado: {resultado}")
        historial.append(f"{numero1} x {numero2} = {resultado}")

    #División    
    elif usuario_operacion == "division" or usuario_operacion == "dividir" or usuario_operacion == "/":
        while not entrada_valida:
            try:
                numero1 = input("Introduce el primer número a dividir")
                numero1 = float(numero1.replace(",", "."))
                numero2 = 0
                while numero2 == 0:
                    numero2 = input("Introduce el segundo número a dividir")
                    numero2 = float(numero2.replace(",", "."))
                    if numero2 == 0:
                        print("No se puede dividir por 0, introduce otro número")
                entrada_valida = True
            except Exception:
                print("Por favor introduzca un número válido")
        
        entrada_valida = False
        resultado = numero1/numero2
        print(f"Resultado: {resultado}")
        historial.append(f"{numero1} / {numero2} = {resultado}")


    #Suma
    elif usuario_operacion == "suma" or usuario_operacion == "sumar" or usuario_operacion == "+":
        while not entrada_valida:
            try:
                numero1 = input("Introduce el primer número a sumar")
                numero1 = float(numero1.replace(",", "."))
                numero2 = input("Introduce el segundo número a sumar")
                numero2 = float(numero2.replace(",", "."))
                entrada_valida = True
            except Exception:
                print("Por favor introduzca un número válido")

        entrada_valida = False
        resultado = numero1+numero2
        print(f"Resultado: {resultado}")
        historial.append(f"{numero1} + {numero2} = {resultado}")


    #Resta
    elif usuario_operacion == "resta" or usuario_operacion == "restar" or usuario_operacion == "-":
        while not entrada_valida:
            try:
                numero1 = input("Introduce el primer número a restar")
                numero1 = float(numero1.replace(",", "."))
                numero2 = input("Introduce el segundo número a restar")
                numero2 = float(numero2.replace(",", "."))
                entrada_valida = True
            except Exception:
                print("Por favor introduzca un número válido")

        entrada_valida = False
        resultado = numero1-numero2
        print(f"Resultado: {resultado}")
        historial.append(f"{numero1} - {numero2} = {resultado}")


    #Potencia
    elif usuario_operacion == "potencia" or usuario_operacion == "xx" or usuario_operacion == "**":
        while not entrada_valida:
            try:
                numero1 = input("Introduce el primer número a potenciar")
                numero1 = float(numero1.replace(",", "."))
                numero2 = input("Introduce el segundo número a potenciar")
                numero2 = float(numero2.replace(",", "."))
                entrada_valida = True
            except Exception:
                print("Por favor introduzca un número válido")

        entrada_valida = False
        resultado = numero1**numero2
        print(f"Resultado: {resultado}")
        historial.append(f"{numero1} ** {numero2} = {resultado}")


    #Salir
    elif usuario_operacion == "salir":
        break

    #Historial
    elif usuario_operacion == "historial":
        print(historial)

    #Otro
    else:
        print("Operación Invalida")
```
